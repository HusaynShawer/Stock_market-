```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stoke_market_info=pd.read_csv("D:\\ML project\\30_yr_stock_market_info.csv")
stoke_market_data=pd.read_csv("D:\\ML project\\30_yr_stock_market_data.csv")
```

```python
stoke_market_data.set_index('Date',inplace=True)
```

```python
stoke_market_info.set_index('SYMBOL',inplace=True)
```

```python
stoke_market_info
```

```python
def plotmarket(x):
    symbol_tybe=x
    print(x)
    symbol=stoke_market_info['SYMBOL_NAME_FULL'][stoke_market_info['SYMBOL_TYPE']==symbol_tybe]
    for i in symbol:
        plt.figure(figsize=(10,6))
        stoke_market_data[i].plot()
        plt.title('{}\n({})'.format(i.upper(),symbol_tybe.lower()))
        plt.xlabel("")
        plt.xticks(rotation=0)
        plt.box(False)
        plt.show()
        
    
```

```python
plotmarket('Market Index')
```

```python
plotmarket('Commodity: AGRICULTURE')
```

```python
plotmarket('Commodity: ENERGY')
```

```python
plotmarket("Commodity: INDUSTRIAL METALS")
```

```python
plotmarket('U.S. Treasury Bonds')
```
